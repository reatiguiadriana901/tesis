```python
import pandas as pd


df = pd.read_csv("datos/morfometria.csv")

df["x"]          = df["x"]          / 10000
df["y"]          = df["y"]          / 10000
df["MBG_Width"]  = df["MBG_Width"]  / 1000
df["MBG_Length"] = df["MBG_Length"] / 1000


df["Hco"]               = df["MAX"] - df["MIN"]
df["Wm"]                = df["MBG_Width"]
df["WM"]                = df["MBG_Length"]
df["Sco"]               = df["Slope_MAX"]
df["Elipticidad"]       = (df["Wm"] / df["WM"]).round(2)
df["Hco_WM"]            = (df["Hco"] / df["WM"]).round(3)
df["Azimut_elongacion"] = df["MBG_Orient"] + 90


resultado = df[["volcan", "x", "y", "Hco", "Wm", "WM", "Sco", "Elipticidad", "Hco_WM", "Azimut_elongacion", "Azimut_de_la_brecha"]].copy()
resultado = resultado.fillna("—")


resultado.columns = [
    "Volcán", "Coordenada X", "Coordenada Y", "Hco (m)", "Wm (m)", "WM (m)", "Sco (°)", 
    "Elipticidad", "Hco/WM", "Az. elongación (°)", "Az. brecha (°)"
]


resultado.to_clipboard(excel=True, index=False)





```
